```python
#Read files we need
bacteria = pd.read_csv('bacteria_KEGG.tsv', sep='\t')
actino = pd.read_csv('actino_KEGG.tsv', sep='\t')
bactero = pd.read_csv('bactero_KEGG.tsv', sep='\t')
balne = pd.read_csv('Balne_KEGG.tsv', sep='\t')
chloro = pd.read_csv('Chloro_KEGG.tsv', sep='\t')
cyano = pd.read_csv('cyano_KEGG.tsv', sep='\t')
eury = pd.read_csv('Eury_KEGG.tsv', sep='\t')
firmi = pd.read_csv('Firmi_KEGG.tsv', sep='\t')
proteo = pd.read_csv('proteo_KEGG.tsv', sep='\t')
```

```python
#From each of the data frames, we generate two new ones: one for the 137 m sample columns and one for the 5-25 m sample columns.
def select_columns(df, bacteria_type):
    if bacteria_type == "137":
        return df.iloc[:, [0, 1]]
    elif bacteria_type == "255":
        return df.iloc[:, [2, 3, 4, 5]]

bacteria137 = select_columns(bacteria, "137")
bacteria255 = select_columns(bacteria, "255")

actino137 = select_columns(actino, "137")
actino255 = select_columns(actino, "255")

bactero137 = select_columns(bactero, "137")
bactero255 = select_columns(bactero, "255")

balne137 = select_columns(balne, "137")
balne255 = select_columns(balne, "255")

chloro137 = select_columns(chloro, "137")
chloro255 = select_columns(chloro, "255")

cyano137 = select_columns(cyano, "137")
cyano255 = select_columns(cyano, "255")

eury137 = select_columns(eury, "137")
eury255 = select_columns(eury, "255")

firmi137 = select_columns(firmi, "137")
firmi255 = select_columns(firmi, "255")

proteo137 = select_columns(proteo, "137")
proteo255 = select_columns(proteo, "255")
```

```python
#Define function to pass all values 0 to 1 and to add ratio column 137
def añadir_ratios137(data):
    #Replace the 0's by 1
    data.replace(0,1, inplace = True)
    # Calcular el ratio 137
    data['ratio_137'] = data['137m_V'] / data['137m_M']    
     # Return the updated DataFrame
    return data

#Define function to pass values from 0 to 1 and to add 25-5 ratio column.
def añadir_ratios255(data):
     #Replace the 0's by 1
    data.replace(0,1, inplace = True)
    # Calculate the average of the 25m and 5m columns for V and M
    data['avg_V'] = (data['25m_V'] + data['5m_V']) / 2
    data['avg_M'] = (data['25m_M'] + data['5m_M']) / 2
    # Calculate the 25-5 ratio
    data['ratio_25_5'] = data['avg_V'] / data['avg_M']
    # Remove temporary columns 'avg_V' and 'avg_M'.
    data.drop(columns=['avg_V', 'avg_M'], inplace=True)
    # Return the updated DataFrame
    return data
```

```python
#We apply the above functions to all files.
#137 to files of 137 and #25 to files of 25:
bacteria137 = añadir_ratios137(bacteria137)
bacteria255 = añadir_ratios255(bacteria255)

actino137 = añadir_ratios137(actino137)
actino255 = añadir_ratios255(actino255) 

bactero137 = añadir_ratios137(bactero137)
bactero255 = añadir_ratios255(bactero255)

balne137 = añadir_ratios137(balne137)
balne255 = añadir_ratios255(balne255)

chloro137 = añadir_ratios137(chloro137)
chloro255 = añadir_ratios255(chloro255)

cyano137 = añadir_ratios137(cyano137)
cyano255 = añadir_ratios255(cyano255)

eury137 = añadir_ratios137(eury137)
eury255 = añadir_ratios255(eury255)

firmi137 = añadir_ratios137(firmi137)
firmi255 = añadir_ratios255(firmi255)

proteo137 = añadir_ratios137(proteo137)
proteo255 = añadir_ratios255(proteo255)
```

```python
#Select only rows with ratios greater than 2 (in ratio 137 or in ratio 25-5).
def filtro2ratio137(data):
    data = data.loc[(data['ratio_137'] >= 2)]
    return data
def filtro2ratio255(data):
    data = data.loc[(data['ratio_25_5'] >= 2)]
    return data

bacteria137filtro2 = filtro2ratio137(bacteria137)
bacteria255filtro2 = filtro2ratio255(bacteria255)

actino137filtro2 = filtro2ratio137(actino137)
actino255filtro2 = filtro2ratio255(actino255)

bactero137filtro2 = filtro2ratio137(bactero137)
bactero255filtro2 = filtro2ratio255(bactero255)

balne137filtro2 = filtro2ratio137(balne137)
balne255filtro2 = filtro2ratio255(balne255)

chloro137filtro2 = filtro2ratio137(chloro137)
chloro255filtro2 = filtro2ratio255(chloro255)

cyano137filtro2 = filtro2ratio137(cyano137)
cyano255filtro2 = filtro2ratio255(cyano255)

eury137filtro2 = filtro2ratio137(eury137)
eury255filtro2 = filtro2ratio255(eury255)

firmi137filtro2 = filtro2ratio137(firmi137)
firmi255filtro2 = filtro2ratio255(firmi255)

proteo137filtro2 = filtro2ratio137(proteo137)
proteo255filtro2 = filtro2ratio255(proteo255)
```

```python
# Define function to sort the above files from highest to lowest ratio and select the first 20 files.
def ordenar_y_seleccionar137(df):
    sorted_df = df.sort_values(by='ratio_137', ascending=False)
    top_20 = sorted_df.head(20)
    return top_20

def ordenar_y_seleccionar255(df):
    sorted_df = df.sort_values(by='ratio_25_5', ascending=False)
    top_20 = sorted_df.head(20)
    return top_20
```

```python
#We order from highest to lowest by ratio and take the top 20.
top20bacteria137 = ordenar_y_seleccionar137(bacteria137)
top20bacteria255 = ordenar_y_seleccionar255(bacteria255)
top20bacteria137.to_csv('top20bacteria137.tsv', sep='\t', index=True)
top20bacteria255.to_csv('top20bacteria255.tsv', sep='\t', index=True)

top20actino137 = ordenar_y_seleccionar137(actino137)
top20actino255 = ordenar_y_seleccionar255(actino255)

top20bactero137 = ordenar_y_seleccionar137(bactero137)
top20bactero255 = ordenar_y_seleccionar255(bactero255)

top20balne137 = ordenar_y_seleccionar137(balne137)
top20balne255 = ordenar_y_seleccionar255(balne255)

top20chloro137 = ordenar_y_seleccionar137(chloro137)
top20chloro255 = ordenar_y_seleccionar255(chloro255)

top20cyano137 = ordenar_y_seleccionar137(cyano137)
top20cyano255 = ordenar_y_seleccionar255(cyano255)

top20eury137 = ordenar_y_seleccionar137(eury137)
top20eury255 = ordenar_y_seleccionar255(eury255)

top20firmi137 = ordenar_y_seleccionar137(firmi137)
top20firmi255 = ordenar_y_seleccionar255(firmi255)

top20proteo137 = ordenar_y_seleccionar137(proteo137)
top20proteo255 = ordenar_y_seleccionar255(proteo255)
```

```python
#We grouped the 20 most abundant KOs in each group and condition into a list, eliminating duplicates.
#137:
dfs137 = {
    'top20actino137': top20actino137,
    'top20bactero137': top20bactero137,
    'top20balne137': top20balne137,
    'top20chloro137': top20chloro137,
    'top20cyano137': top20cyano137,
    'top20eury137': top20eury137,
    'top20firmi137': top20firmi137,
    'top20proteo137': top20proteo137
}

kos137 = []

for df_name, df in dfs137.items():
    indices = df.index.tolist()
    kos137 += indices  # Utilizar += en lugar de append() para agregar índices directamente a la lista

kos137 = list(set(kos137))

#25-5:
dfs255 = {
    'top20actino255': top20actino255,
    'top20bactero255': top20bactero255,
    'top20balne255': top20balne255,
    'top20chloro255': top20chloro255,
    'top20cyano255': top20cyano255,
    'top20eury255': top20eury255,
    'top20firmi255': top20firmi255,
    'top20proteo255': top20proteo255
}

kos255 = []

for df_name, df in dfs255.items():
    indices = df.index.tolist()
    kos255 += indices  # Utilizar += en lugar de append() para agregar índices directamente a la lista

kos255 = list(set(kos255))
```

```python
#We extract the orthologous KEGGs common to both depths
kos137_set = set(kos137)
kos255_set = set(kos255)

intersection_kos = kos137_set.intersection(kos255_set)

print(intersection_kos)
```

```python
#KOs list description annotation function:
def annotate_KEGG_KOs(kos):
    KEGG_anot = open("KEGG_KO", "r")  #  KEGGs database
    KO_dict = {}
    for line in KEGG_anot:
        Ko, description = line.rstrip().split("\t")
        Ko = Ko.rstrip()
        KO_dict[Ko] = [description]
    
    results = []
    for KEGG in kos:
        if KEGG in KO_dict:
            results.append([KEGG, KO_dict[KEGG][0].rstrip()])
    
    df = pd.DataFrame(results, columns=["KO", "Description"])
    return df                

#137:
dffilos137 = annotate_KEGG_KOs(kos137)

#25-5:
dffilos255 = annotate_KEGG_KOs(kos255)

#Intersection
intersecciondescription = annotate_KEGG_KOs(intersection_kos)
```

```python
#For each of the conditions, the most abundant KOs in that condition, their description and the ratios in each of the blades, are listed in the same table.
#137:
actino137 = actino137.rename_axis('KO').reset_index()
actino137 = actino137[['KO', 'ratio_137']]
dfheat1 = pd.merge(dffilos137, actino137, on='KO', how='left')
dfheat1 = dfheat1.rename(columns={'ratio_137': 'Actinobacteria'})
dfheat1['Actinobacteria'] = dfheat1['Actinobacteria'].fillna(0)

bactero137 = bactero137.rename_axis('KO').reset_index()
bactero137 = bactero137[['KO', 'ratio_137']]
dfheat2 = pd.merge(dfheat1, bactero137, on='KO', how='left')
dfheat2 = dfheat2.rename(columns={'ratio_137': 'Bacteroidetes'})
dfheat2['Bacteroidetes'] = dfheat2['Bacteroidetes'].fillna(0)

balne137 = balne137.rename_axis('KO').reset_index()
balne137 = balne137[['KO', 'ratio_137']]
dfheat3 = pd.merge(dfheat2, balne137, on='KO', how='left')
dfheat3 = dfheat3.rename(columns={'ratio_137': 'Balneolaeota'})
dfheat3['Balneolaeota'] = dfheat3['Balneolaeota'].fillna(0)

chloro137 = chloro137.rename_axis('KO').reset_index()
chloro137 = chloro137[['KO', 'ratio_137']]
dfheat4 = pd.merge(dfheat3, chloro137, on='KO', how='left')
dfheat4 = dfheat4.rename(columns={'ratio_137': 'Chloroflexi'})
dfheat4['Chloroflexi'] = dfheat4['Chloroflexi'].fillna(0)


cyano137 = cyano137.rename_axis('KO').reset_index()
cyano137 = cyano137[['KO', 'ratio_137']]
dfheat5 = pd.merge(dfheat4, cyano137, on='KO', how='left')
dfheat5 = dfheat5.rename(columns={'ratio_137': 'Cyanobacteria'})
dfheat5['Cyanobacteria'] = dfheat5['Cyanobacteria'].fillna(0)


eury137 = eury137.rename_axis('KO').reset_index()
eury137 = eury137[['KO', 'ratio_137']]
dfheat6 = pd.merge(dfheat5, eury137, on='KO', how='left')
dfheat6 = dfheat6.rename(columns={'ratio_137': 'Euryarchaeota'})
dfheat6['Euryarchaeota'] = dfheat6['Euryarchaeota'].fillna(0)


firmi137 = firmi137.rename_axis('KO').reset_index()
firmi137 = firmi137[['KO', 'ratio_137']]
dfheat7 = pd.merge(dfheat6, firmi137, on='KO', how='left')
dfheat7 = dfheat7.rename(columns={'ratio_137': 'Firmicutes'})
dfheat7['Firmicutes'] = dfheat7['Firmicutes'].fillna(0)



proteo137 = proteo137.rename_axis('KO').reset_index()
proteo137 = proteo137[['KO', 'ratio_137']]
dfheat8 = pd.merge(dfheat7, proteo137, on='KO', how='left')
dfheat8 = dfheat8.rename(columns={'ratio_137': 'Proteobacteria'})
dfheat8['Proteobacteria'] = dfheat8['Proteobacteria'].fillna(0)

```
